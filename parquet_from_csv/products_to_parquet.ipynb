```python
import pandas as pd
import numpy as np
import plotly.express as px
import os
import matplotlib.pyplot as plt
from numpy import int8

src = 'finalCCRSdata/'
```

```python
file_list = os.listdir(src)

# get a list of prefixes
# prefix_list = list(set([file.split('_')[0] for file in file_list if '_' in file and file.endswith('.csv')]))
prefix_list = ['Product']


# make a dictionary of dataframes
for prefix in prefix_list:
    prefix_files = [file for file in file_list if file.startswith(prefix) and file.endswith('.csv')]
    df_list = [pd.read_csv(os.path.join(src, file)) for file in prefix_files]
    df_concat = pd.concat(df_list, keys=[(prefix, file) for file in prefix_files], ignore_index=True)
```

```python
df_concat.columns
```

```python
df_concat = df_concat[['ProductId', 'LicenseeId', 'InventoryType', 'Name', 'Description',
                       'UnitWeightGrams', 'ExternalIdentifier', 'IsDeleted',
                       #'CreatedBy',
                       'CreatedDate', 
                       #'UpdatedBy', 'UpdatedDate'
              ]]
#del df_list
#del file_list
#del df_concat
```

```python
df_concat = df_concat.astype({
'ProductId' :                int,
'LicenseeId' :               int,
'InventoryType' :           'category',
'Name' :                    str,
'Description' :             str,
'UnitWeightGrams' :        float,
'ExternalIdentifier' :      str,
'IsDeleted' :                 bool,
'CreatedDate' :             object,
}, errors='ignore')
df_concat['CreatedDate'] = pd.to_datetime(df_concat['CreatedDate'])
```

```python
df_concat.to_parquet('products.parquet',index=False)
```
